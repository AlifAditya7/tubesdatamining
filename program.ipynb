### Data Preprocessing

```python
# mengecek data duplikat
duplicates = data.duplicated()
data.drop_duplicates(inplace=True)

# mengecek data kosong
empty = data.isnull()
data.dropna(inplace=True)

data.head(10)
```

```python
# cek data tidak relevan
# Ganti '#NUM!' dengan NaN
data = data.replace('#NUM!', np.nan)

# Pastikan untuk mengonversi semua kolom ke tipe yang benar
data = data.astype(float)

# Sekarang Anda dapat menghitung korelasinya
correlations = data.corr()['is_safe'].sort_values()

# fitur dengan korelasi absolut kurang dari ambang batas 1,0001 sebagai tidak relevan
irrelevant_features = correlations[np.abs(correlations) < 1.0001]

print(irrelevant_features)
```

### Data Transformation

```python
# Transformasi data dengan normalisasi min-max
# Kolom yang akan di normalisasi
columns = ['aluminium', 'ammonia', 'arsenic', 'barium', 'cadmium', 'chloramine', 'chromium', 'copper', 'flouride', 'bacteria', 'viruses', 'lead', 'nitrates', 'nitrites', 'mercury', 'perchlorate', 'radium', 'selenium', 'silver', 'uranium']
columns1 = ['aluminium', 'ammonia', 'arsenic', 'barium', 'cadmium', 'chloramine', 'chromium', 'copper', 'flouride', 'bacteria', 'viruses', 'lead', 'nitrates', 'nitrites', 'mercury', 'perchlorate', 'radium', 'selenium', 'silver', 'uranium','is_safe']

# Melakukan normalisasi Min-Max hanya untuk 10 data pertama
num_data = 10
normalized = (data[columns].head(num_data) - data[columns].min()) / (data[columns].max() - data[columns].min())
normalized1 = (data[columns1].head(num_data) - data[columns1].min()) / (data[columns1].max() - data[columns1].min())

# Tampilkan hasil normalisasi
print(normalized)
```

### Menghitung Jarak Antar Data Latih Menggunakan Euclidean Distance

```python
# Konversi Pandas DataFrame ke array NumPy
data_array = normalized.to_numpy()

# Atur ukuran batch
batch_size = 10

# Get the number of rows in the data
num_rows = data_array.shape[0]

# Dapatkan jumlah baris dalam data
distances1 = np.zeros((num_rows, num_rows))

# Hitung jarak dalam batch
for i in range(0, num_rows, batch_size):
    start_idx = i
    end_idx = min(i + batch_size, num_rows)
    
    # Menghitung jarak Euclidean antar data
    distances1[start_idx:end_idx, :] = np.linalg.norm(data_array[start_idx:end_idx, np.newaxis, :] - data_array, axis=2)

# Cetak matriks jarak
print(distances1)
```
