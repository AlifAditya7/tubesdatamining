```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('waterQuality1.csv')
pd.set_option('display.max_columns', None)
data.info()
```

```python
# menampilkan data awal 
data.head(10)
```

```python
# mengecek data duplikat
duplicates = data.duplicated()
data.drop_duplicates(inplace=True)

# mengecek data kosong
empty = data.isnull()
data.dropna(inplace=True)

data.head(10)
```

```python
# Ganti '#NUM!' dengan NaN
data = data.replace('#NUM!', np.nan)

# Pastikan untuk mengonversi semua kolom ke tipe yang benar
data = data.astype(float)

# Sekarang Anda dapat menghitung korelasinya
correlations = data.corr()['is_safe'].sort_values()

# Anggap fitur dengan korelasi absolut kurang dari ambang batas (misalnya 1,0001) sebagai tidak relevan
irrelevant_features = correlations[np.abs(correlations) < 1.0001]

print(irrelevant_features)
```

```python
# cek data tidak relevan
# data transformasi, melakukan normalisasi min-max
columns = ['aluminium', 'ammonia', 'arsenic', 'barium', 'cadmium', 'chloramine', 'chromium', 'copper', 'flouride', 'bacteria', 'viruses', 'lead', 'nitrates', 'nitrites', 'mercury', 'perchlorate', 'radium', 'selenium', 'silver', 'uranium', 'is_safe']

# Mengubah semua kolom menjadi numerik
for col in columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Melakukan normalisasi Min-Max
normalized = (data - data.min()) / (data.max() - data.min())

print(normalized)
```
